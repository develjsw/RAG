```python
# 환경변수 로드
from dotenv import load_dotenv

load_dotenv()
```

```python
# LangChain기반 LLM 설정
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
llm
```

```python
# LLM 호출
ai_msg = llm.invoke("대한민국 정부가 세워진 날은 언제인가요?")
```
