# LCEL (LangChain Expression Language)

**파이프(`|`) 연산자로 컴포넌트를 연결해 체인을 구성하는 방식 (python 기준)**<br>
`prompt | llm | parser`처럼 좌→우로 데이터가 흐름

- 흐름: 프롬프트 → LLM → 출력 파서(`StrOutputParser`)
- 핵심: `|`로 이으면 하나의 `RunnableSequence`(= Runnable)가 됨 → `invoke`(단일)·`batch`(다중)·`stream`(스트리밍) 등으로 실행
- ※ LCEL은 별도 문법이 아니라 `Runnable` 합성을 부르는 이름 (v1 docs엔 LCEL 개념 전용 페이지가 없고 `Runnable` 인터페이스로 통합됨)

# 2. LCEL 체인 구성 및 실행

- `ChatPromptTemplate` → `ChatOpenAI` → `StrOutputParser`를 `|`로 연결
- `StrOutputParser` = LLM 응답(`AIMessage`)에서 문자열 본문만 추출
- `chain.invoke({...})` = 템플릿 변수(`topic`)를 채워 체인 실행

```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

prompt = ChatPromptTemplate.from_template(
    "{topic}에 대해 백엔드 개발자 관점에서 한 문장으로 설명해줘"
)

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

parser = StrOutputParser()

chain = prompt | llm | parser

result = chain.invoke({
    "topic": "랭체인"
})
```
